```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Seleccionar las columnas 'Total Cup Points' y 'Flavor' del DataFrame df
X = df[['Total Cup Points', 'Flavor']]

# Especificar el número de clusters deseado
n_clusters = 3

# Crear una instancia del modelo K-means
kmeans = KMeans(n_clusters=n_clusters)

# Ajustar el modelo a los datos
kmeans.fit(X)

# Obtener las etiquetas de los clusters
labels = kmeans.labels_

# Obtener las coordenadas de los centroides
centroids = kmeans.cluster_centers_

# Visualizar los datos y los clusters
plt.scatter(X['Total Cup Points'], X['Flavor'], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', c='red', s=200)
plt.xlabel('Total Cup Points')
plt.ylabel('Flavor')
plt.title('K-means Clustering: Total Cup Points vs Flavor')
plt.show()
```
